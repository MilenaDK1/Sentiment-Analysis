```python
import sys
import os
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.datasets import load_svmlight_file
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report

cwd = os.getcwd()

for VectorSize in [100, 300]:
		for WindowSize in [5, 15]:
			for Metric in [1, 2]:
				
				prefix = f"v{VectorSize}_w{WindowSize}_m{Metric}"
				
				path = cwd+'/train/train_'+prefix+'.txt'
				print('Carregando '+path)
				X_train, y_train = load_svmlight_file(path)

				path = cwd+'/test/test_'+prefix+'.txt'
				print('Carregando '+path)
				X_test, y_test = load_svmlight_file(path)

				clf = make_pipeline(StandardScaler(with_mean=False), SVC(gamma='auto'))
				print('Treinando...')
				clf.fit(X_train, y_train)
				print('Testando...')
				y_pred = clf.predict(X_test)

				print ('Accuracy: ', clf.score(X_test, y_test))

				# cria a matriz de confusao
				cm = confusion_matrix(y_test, y_pred)
				print (cm)
				print(classification_report(y_test, y_pred))
				print()

 

```
